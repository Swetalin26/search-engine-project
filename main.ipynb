```python
# Stopword Removal
STOPWORDS = {
    "a","an","the","is","are","was","were","be","been","have","has","had",
    "do","does","did","will","would","could","should","and","or","but",
    "in","on","at","to","for","of","with","by","from","as","that","this",
    "it","its","they","them","we","our","you","he","she","all","not",
    "also","can","use","used","which","where","when","how","what","who"
}

print(f"Stopwords loaded: {len(STOPWORDS)} words will be ignored")
```

```python
#  Tokenization
#  Basic Stemming 

def tokenize(text):
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', ' ', text)   # remove punctuation
    words = text.split()
    words = [w for w in words if w not in STOPWORDS and len(w) > 1]
    return words

def stem(word):
    # Basic suffix stripping (manual stemmer)
    for suffix in ['ing', 'tion', 'ness', 'ment', 'ers', 'es', 'ed', 's']:
        if word.endswith(suffix) and len(word) - len(suffix) > 2:
            return word[:-len(suffix)]
    return word

def tokenize_stemmed(text):
    tokens = tokenize(text)
    return [stem(w) for w in tokens]

# Test it
print("Normal  :", tokenize("Machine learning is running fast"))
print("Stemmed :", tokenize_stemmed("Machine learning is running fast"))
```

```python
#  Inverted Index
# Maps every word → list of document IDs that contain it

def build_index(database):
    index = {}
    for doc in database:
        text = doc["title"] + " " + doc["content"]
        tokens = set(tokenize_stemmed(text))   # set = no duplicates
        for token in tokens:
            if token not in index:
                index[token] = []
            index[token].append(doc["id"])
    return index

index = build_index(database)
print(f" Inverted index built: {len(index)} unique terms indexed")
print(f" Sample → 'machin': {index.get('machin', [])}")
print(f" Sample → 'climat': {index.get('climat', [])}")
```

```python
# TF-IDF Ranking

def compute_tfidf(query_tokens, database):
    N = len(database)
    scores = {}

    for doc in database:
        title_tokens = tokenize_stemmed(doc["title"])
        body_tokens  = tokenize_stemmed(doc["content"])

        # Title gets 2x weight boost
        all_tokens = title_tokens * 2 + body_tokens

        score = 0
        for term in query_tokens:
            # TF: how often term appears in this document
            tf = all_tokens.count(term) / len(all_tokens) if all_tokens else 0

            # IDF: how rare is this term across all documents
            docs_with_term = sum(
                1 for d in database
                if term in tokenize_stemmed(d["title"] + " " + d["content"])
            )
            idf = math.log(N / (docs_with_term + 1)) + 1

            score += tf * idf

        scores[doc["id"]] = score

    # Sort by score highest first
    ranked = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    return [(doc_id, score) for doc_id, score in ranked if score > 0]

print("TF-IDF scorer ready (with title 2x boost)")
```

```python
#  Category Filter
#  Boolean Search (AND / OR / NOT)

def apply_boolean(query, database):
    if " AND " in query.upper():
        parts = query.upper().split(" AND ")
        original_parts = query.split(" AND " if " AND " in query else " and ")
        results = database
        for part in original_parts:
            tokens = tokenize_stemmed(part.strip())
            results = [d for d in results if any(
                t in tokenize_stemmed(d["title"] + " " + d["content"]) for t in tokens
            )]
        return results, tokenize_stemmed(query.replace(" AND ","").replace(" and ",""))

    elif " OR " in query.upper():
        parts = query.split(" OR " if " OR " in query else " or ")
        results = []
        all_tokens = []
        for part in parts:
            tokens = tokenize_stemmed(part.strip())
            all_tokens.extend(tokens)
            results += [d for d in database if any(
                t in tokenize_stemmed(d["title"] + " " + d["content"]) for t in tokens
            )]
        seen = set()
        results = [d for d in results if not (d["id"] in seen or seen.add(d["id"]))]
        return results, all_tokens

    elif " NOT " in query.upper():
        parts = query.split(" NOT " if " NOT " in query else " not ")
        include_tokens = tokenize_stemmed(parts[0].strip())
        exclude_tokens = tokenize_stemmed(parts[1].strip())
        results = [d for d in database if not any(
            t in tokenize_stemmed(d["title"] + " " + d["content"]) for t in exclude_tokens
        )]
        return results, include_tokens

    return database, tokenize_stemmed(query)

def filter_by_category(database, category=None):
    if not category or category.lower() == "all":
        return database
    return [d for d in database if d["category"].lower() == category.lower()]

print("Boolean search ready (AND / OR / NOT)")
print("Category filter ready (Technology / Science / History / Health / Business)")
```

```python
# Feature 9: Autocomplete Suggestions

all_words = set()
for doc in database:
    tokens = tokenize(doc["title"] + " " + doc["content"])
    all_words.update(tokens)

def autocomplete(prefix):
    prefix = prefix.lower()
    suggestions = sorted([w for w in all_words if w.startswith(prefix)])
    return suggestions[:5]   # top 5 suggestions

# Test it
print("Autocomplete 'mac' →", autocomplete("mac"))
print("Autocomplete 'cli' →", autocomplete("cli"))
print("Autocomplete 'vac' →", autocomplete("vac"))
```

```python
# Search History

search_history = []

def add_to_history(query, result_count):
    search_history.append({"query": query, "results": result_count})

def show_history():
    if not search_history:
        print("No search history yet.")
        return
    print("\nSearch History:")
    print("-" * 35)
    for i, h in enumerate(search_history, 1):
        print(f"  {i}. '{h['query']}' → {h['results']} result(s)")

print("Search history tracker ready")
```

```python
# Puts ALL features together

def search(query, database, category=None, top_n=5):

    print(f"\nSearching for: '{query}'")

    # Autocomplete hint
    first_word = query.split()[0] if query.split() else ""
    suggestions = autocomplete(first_word)
    if suggestions:
        print(f"  Suggestions: {', '.join(suggestions)}")

    # Apply boolean logic
    pool, query_tokens = apply_boolean(query, database)

    # Apply category filter
    pool = filter_by_category(pool, category)

    if not query_tokens:
        print("Please enter a valid query.")
        return

    # Compute TF-IDF scores
    results = compute_tfidf(query_tokens, pool)

    # Save to history
    add_to_history(query, len(results))

    if not results:
        print(f" No results found for '{query}'")
        print(f"   Try: machine learning, vaccines, napoleon, bitcoin")
        return

    print(f"{len(results)} result(s) found\n")
    print("=" * 55)

    for rank, (doc_id, score) in enumerate(results[:top_n], 1):
        doc = next(d for d in database if d["id"] == doc_id)

        # Score bar
        bar_len = int((score / results[0][1]) * 15)
        bar = "|" * bar_len + "*" * (15 - bar_len)

        print(f"#{rank}  {doc['title']}")
        print(f"     Category : {doc['category']}")
        print(f"     Score    : {score:.4f}  [{bar}]")
        print(f"     Preview  : {doc['content'][:100]}...")
        print()

    print("=" * 55)

print("Search engine fully ready! All 10 features loaded.")
print()
print("Features included:")
print("  1. Stopword Removal")
print("  2. Tokenization")
print("  3. Stemming")
print("  4. Inverted Index")
print("  5. TF-IDF Ranking")
print("  6. Title Boost (2x)")
print("  7. Category Filter")
print("  8. Boolean Search (AND/OR/NOT)")
print("  9. Autocomplete")
print(" 10. Search History")
```
