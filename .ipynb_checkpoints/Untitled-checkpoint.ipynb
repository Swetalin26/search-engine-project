```python
import re

def tokenize(text):
    text = text.lower()                        # "Machine" → "machine"
    text = re.sub(r'[^a-z0-9\s]', ' ', text)  # remove punctuation
    words = text.split()                       # split by spaces
    return words

# Try it:
print(tokenize("Machine Learning is great!"))
# Output: ['machine', 'learning', 'is', 'great']
```

```python
STOPWORDS = {"a", "an", "the", "is", "are", "was", "in", "of", "and", "or", "to", "for"}

def clean_tokens(words):
    return [w for w in words if w not in STOPWORDS and len(w) > 1]

# Try it:
tokens = tokenize("Machine Learning is a subset of AI")
print(clean_tokens(tokens))
```

```python
from collections import defaultdict

def build_index(database):
    index = defaultdict(list)   # word → [list of doc ids]

    for doc in database:
        text = doc["title"] + " " + doc["content"]
        tokens = clean_tokens(tokenize(text))

        for word in set(tokens):          # set() avoids adding same doc twice
            index[word].append(doc["id"])

    return index

# Try it:
index = build_index(database)
print(index["machine"])   # → [1, 5]  (doc 1 and 5 contain "machine")
```

```python
import math

def compute_tfidf(query_tokens, database):
    N = len(database)     # total number of documents
    scores = {}           # doc_id → score

    for doc in database:
        doc_id = doc["id"]
        text = doc["title"] + " " + doc["content"]
        tokens = clean_tokens(tokenize(text))
        scores[doc_id] = 0

        for term in query_tokens:
            # TF = count of term in this doc / total words in doc
            tf = tokens.count(term) / len(tokens) if tokens else 0

            # IDF = log(total docs / docs containing this term)
            docs_with_term = sum(1 for d in database
                                 if term in clean_tokens(tokenize(d["title"] + " " + d["content"])))
            idf = math.log(N / (docs_with_term + 1)) + 1

            scores[doc_id] += tf * idf

    # Sort by score, highest first
    ranked = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    return [(doc_id, score) for doc_id, score in ranked if score > 0]
```

```python
def search(query, database):
    query_tokens = clean_tokens(tokenize(query))

    if not query_tokens:
        print("Please enter a valid query.")
        return

    results = compute_tfidf(query_tokens, database)

    if not results:
        print("No results found.")
        return

    print(f"\n{len(results)} result(s) found:\n")
    for rank, (doc_id, score) in enumerate(results, 1):
        doc = next(d for d in database if d["id"] == doc_id)
        print(f"{rank}. {doc['title']}")
        print(f"   Score: {score:.4f}")
        print(f"   {doc['content'][:100]}...")
        print()
```

```python
def compute_tfidf(query_tokens, database):
    for doc in database:
        title_tokens = tokenize(doc["title"])
        content_tokens = tokenize(doc["content"])

        title_score = 0
        content_score = 0

        for term in query_tokens:
            # title match is worth 3x more than content match
            title_score  += title_tokens.count(term) * 3
            content_score += content_tokens.count(term) * 1

        doc["score"] = title_score + content_score
```

```python
import math

STOPWORDS = {"a","an","the","is","are","was","were","be","been","have","has","had",
             "do","does","did","will","would","could","should","and","or","but",
             "in","on","at","to","for","of","with","by","from","as","that","this",
             "it","they","we","you","he","she","all","not","its"}

def tokenize(text):
    import re
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', ' ', text)
    words = text.split()
    return [w for w in words if w not in STOPWORDS and len(w) > 1]

def search(query, database, top_n=5):
    query_tokens = tokenize(query)
    
    if not query_tokens:
        print("Please type a proper query!")
        return
    
    N = len(database)
    scores = {}

    for doc in database:
        tokens = tokenize(doc["title"] + " " + doc["content"])
        score = 0
        for term in query_tokens:
            tf = tokens.count(term) / len(tokens)
            df = sum(1 for d in database if term in tokenize(d["title"] + " " + d["content"]))
            idf = math.log(N / (df + 1)) + 1
            score += tf * idf
        scores[doc["id"]] = score

    ranked = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    results = [(doc_id, score) for doc_id, score in ranked if score > 0]

    if not results:
        print(f"No results found for '{query}'")
        return

    print(f"\n Results for: '{query}'")
    print(f" {len(results)} match(es) found\n")
    print("-" * 50)

    for rank, (doc_id, score) in enumerate(results[:top_n], 1):
        doc = next(d for d in database if d["id"] == doc_id)
        print(f"#{rank}  {doc['title']}")
        print(f"    Category : {doc['category']}")
        print(f"    Score    : {score:.4f}")
        print(f"    Preview  : {doc['content'][:100]}...")
        print()

print("Search function ready!")
```
